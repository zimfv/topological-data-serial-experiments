```python
n = 24
experiments = 10

dfs_euler = []
fc = get_hexagonal_tessolation_on_torus2d(n)
for i in range(experiments):
    fc.set_filtration(np.random.random(n**2))
    fc.initialize_complex()
    dfs_euler.append(fc.persistent_euler())
```

```python
def calculate_mean_perisstent_euler(dfs):
    vals = np.unique(np.concatenate([df[['A', 'B']].values for df in dfs]))
    df = pd.DataFrame({'A': vals[:-1], 'B': vals[1:]})
    df['C'] = 0.5*df['A'] + 0.5*df['B']
    df.loc[df['C'] == -np.inf, 'C'] = df.loc[df['C'] == -np.inf, 'B'] - 1
    df.loc[df['C'] == +np.inf, 'C'] = df.loc[df['C'] == +np.inf, 'A'] + 1
    df['chars'] = [[i[(i['A'] < c) & (i['B'] > c)]['Euler Char'].values[0] for i in dfs] for c in df['C']]
    df['Mean Euler Char'] = [np.mean(i) for i in df['chars']]
    return df[['A', 'B', 'Mean Euler Char']]
```

```python
mpe = calculate_mean_perisstent_euler(dfs_euler)
mpe
```

```python
def df_func(array, df, colname='Mean Euler Char'):
    # 
    return [df[(df['A'] <= x) & (df['B'] > x)][colname].values[0] for x in array]
```

```python
x = np.arange(0, 1.001, 0.001)
y = df_func(x, mpe)

plt.figure(figsize=(12, 8))
plt.title('Emperic Euler Charecteristics for splitting the torus by {0}x{0} square cells'.format(n))
plt.plot(x, y, color='indigo', linewidth=2)
plt.grid(True)
plt.show()
```
