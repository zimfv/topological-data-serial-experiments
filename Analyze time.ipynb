```python
import time
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
from matplotlib import colormaps as cm
from package.tessellations import *
```

# get_cubical_tessellation_on_torus2d

```python
ns = np.arange(0, 101, 4)
ns[ns==0] = 2
times = {'get_cubical_tessellation_on_torus2d' : [], 
         'set_filtration' : [],
         'initialize_complex' : [],
         'get_cycle_info' : [], 
         'get_giant_cycles' : []}

bar = tqdm(total=len(ns))
for n in ns:
    # get_cubical_tessellation_on_torus2d
    clock = time.perf_counter()
    fc = get_cubical_tessellation_on_torus2d(n)
    times['get_cubical_tessellation_on_torus2d'].append(time.perf_counter() - clock)
    # set_filtration
    filtration = np.random.random(n**2)
    clock = time.perf_counter()
    fc.set_filtration(filtration)
    times['set_filtration'].append(time.perf_counter() - clock)
    # initialize_complex
    clock = time.perf_counter()
    fc.initialize_complex()
    times['initialize_complex'].append(time.perf_counter() - clock)
    # get_cycle_info
    clock = time.perf_counter()
    fc.get_cycle_info()
    times['get_cycle_info'].append(time.perf_counter() - clock)
    # get_giant_cycles
    clock = time.perf_counter()
    fc.get_giant_cycles()
    times['get_giant_cycles'].append(time.perf_counter() - clock)
    
    bar.update()
bar.close()

for label in times:
    times[label] = np.array(times[label])
```

```python
colormap = cm['turbo']
n = len(times)
colornums = np.arange(n) / (n-1)
colors = {}
for i in range(n):
    colors.update({list(times.keys())[i] : colormap(i/(n-1))})
colors
```

```python
plt.figure(figsize=(16, 12))
for label in times:
    plt.plot(ns, times[label], linewidth=4, label=label, color=colors[label])
    plt.scatter(ns, times[label], linewidth=3, color=colors[label])
plt.grid(True)
plt.legend()
plt.show()
```
