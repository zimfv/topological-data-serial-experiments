```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from package.tessellations import get_cubical_tessellation_on_torus2d, get_hexagonal_tessolation_on_torus2d
from itertools import product
```

```python
figsize = (12, 12)
```

```python
filename = 'LaTeX/pics/Voronoi{0}.png'
```

```python
n = 6
filtration = np.random.random(size=n**2)
obj = get_cubical_tessellation_on_torus2d(n, filtration=filtration)

plt.figure(figsize=figsize)
obj.draw_filtration(colormap='rainbow', edgecolor='black')

points = np.array(list(product(np.arange(6) + 0.5, repeat=2)))
plt.scatter(points[:, 0], points[:, 1], color='black', linewidth=4)
plt.xticks([])
plt.yticks([])
plt.savefig(filename.format('Z'))
```

```python
n = 6
filtration = np.random.random(size=n**2)
obj = get_hexagonal_tessolation_on_torus2d(n, filtration=filtration)

plt.figure(figsize=figsize)
obj.draw_filtration(colormap='rainbow', edgecolor='black')

points = np.array(obj.cords).mean(axis=1)

plt.scatter(points[:, 0], points[:, 1], color='black', linewidth=4)
plt.xticks([])
plt.yticks([])
plt.savefig(filename.format('A'))
```

```python
def tosqrt3(i):
    if i%1 == 0.5:
        s = str(i)[:str(i).find('.')]
        s = '$' + s + '\\frac{\sqrt{3}}{2}$'
    else:
        s = str(i)[:str(i).find('.')]
    return s
```

```python
filename = 'LaTeX/pics/percolationBlues{0}.png'

n = 6
filtration = np.random.random(size=n**2)
obj = get_hexagonal_tessolation_on_torus2d(n, filtration=filtration)

plt.figure(figsize=figsize)
obj.draw_filtration(colormap='Blues', edgecolor='orangered')

points = np.array(obj.cords).mean(axis=1)


xticks = np.arange(-1, 2*n+1)
xlabels = xticks%(2*n) / 2
plt.xticks(xticks, labels=xlabels)

yticks = np.arange(-1, 1.5*n, 0.5)
ylabels = yticks % (1.5*n)
ylabels = [tosqrt3(i) for i in ylabels]
print(ylabels)
plt.yticks(yticks, labels=ylabels)

plt.savefig(filename.format('A'))
```
