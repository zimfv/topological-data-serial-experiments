```python
import time
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import colormaps as cm
```

# Analyze lab_cubical_tessellation_on_torus2d_modern

```python
sizes = np.arange(3, 129)
experiments = 100
filename = 'lab-results/lab_cubical_tessellation_on_torus2d_modern_aggregated/size{0}.csv'
dfs = [pd.read_csv(filename.format(size)) for size in sizes]
dfs = [df[df['Death'] == np.inf][['Birth', 'Dimension', 'Space']] for df in dfs]
for i in range(len(sizes)):
    dfs[i]['Size'] = sizes[i]
df = pd.concat(dfs)
df['Number'] = np.tile(np.arange(3), len(df)//3)
df
```

```python
df_mean = df.groupby(['Size', 'Number'], as_index=False).mean()[['Size', 'Number', 'Birth']]
ks = [df_mean[df_mean['Number'] == i]['Birth'].values for i in range(3)]

plt.figure(figsize=(12, 6))
plt.plot(sizes, ks[2], linewidth=4, color='steelblue', label='Mean 2nd giant 1-cycle birth$')
plt.plot(sizes, ks[1], linewidth=4, color='indigo', label='Mean 1st giant 1-cycle birth $k_1^{perc}$')
plt.legend()

xticks = np.unique(np.concatenate([[np.min(sizes), np.max(sizes)], sizes[np.arange(0, len(sizes), len(sizes)//10)]]))
plt.xticks(xticks)
plt.grid(True)
plt.savefig('LaTeX/pics/limitZ2.png')
```

# Analyze lab_cubical_tessellation_on_torus3d_aggregated

```python
sizes = np.arange(3, 70)
experiments = 100
filename = 'lab-results/lab_cubical_tessellation_on_torus3d_aggregated/size{0}.csv'
clock = time.perf_counter()
dfs = [pd.read_csv(filename.format(size)) for size in sizes]
print('Loaded in {0: .4f} seconds.'.format(time.perf_counter() - clock))
for i in range(len(sizes)):
    dfs[i]['Size'] = sizes[i]
```

```python
df = [i[i['Death'] == np.inf] for i in dfs]
df = pd.concat(df)
df['Number'] = np.tile(np.arange(7), len(df)//7)
df.head(16)
```

```python
assert (df['Dimension'].values == np.tile([0, 1, 1, 1, 2, 2, 2], len(df)//7)).all()
```

```python
df_mean = df.groupby(['Size', 'Number'], as_index=False).mean()[['Size', 'Number', 'Birth']]
ks = np.array([df_mean[df_mean['Number'] == i]['Birth'].values for i in range(7)])
```

```python
births1 = ks[1:4]
births2 = ks[4:]
colors1 = [cm['winter'](i/2) for i in range(3)]
colors2 = [cm['autumn'](i/2) for i in range(3)]
labels1 = ['Mean 1st giant 1-cycle birth $k_1^{perc}$', 
           'Mean 2nd giant 1-cycle birth', 
           'Mean 3d giant 1-cycle birth']
labels2 = ['Mean 1st giant 2-cycle birth $k_2^{perc}$', 
           'Mean 2nd giant 2-cycle birth', 
           'Mean 3d giant 2-cycle birth']


plt.figure(figsize=(12, 5))
for i in range(3):
    plt.plot(sizes, births1[i], color=colors1[i], label=labels1[i], linewidth=4)
for i in range(3):
    plt.plot(sizes, births2[i], color=colors2[i], label=labels2[i], linewidth=4)

xticks = np.unique(np.concatenate([[np.min(sizes), np.max(sizes)], sizes[np.arange(0, len(sizes), len(sizes)//10)]]))
plt.xticks(xticks)
plt.legend()
plt.grid(True)
plt.savefig('LaTeX/pics/limitZ3.png')
```
