# General position

```python
def get_triangle_radius(a, b, c, norm=np.linalg.norm):
    """
    Returns radius of circumscribed circle of triangle.
    
    Parameters:
    -------
    a : np.ndarray
        Vertex
        
    b : np.ndarray
        Vertex
        
    c : np.ndarray
        Vertex
    
    norm : function
        Matrix or vector norm.
    
    Returns:
    --------
    r : float
        Radius.
    """
    al = norm(a-b)
    bl = norm(b-c)
    cl = norm(c-a)
    p = 0.5*(al+bl+cl)
    top = al*bl*cl
    bot = 4 * np.sqrt(p*(p-al)*(p-bl)*(p-cl))
    r = top/bot
    return r
```

```python
def points_on_line(a, b, c, norm=np.linalg.norm):
    """
    Check does 3 points lie on one line.
    
    Parameters:
    -----------
    a : float np.ndarray
    
    b : float np.ndarray
    
    c : float np.ndarray
    
    norm : function
        Matrix or vector norm.
    
    Returns:
    g : bool
    """
    ab = b - a
    bc = c - b
    ca = a - c
    
    abl = np.linalg.norm(ab)
    bcl = np.linalg.norm(bc)
    cal = np.linalg.norm(ca)
    
    g = (abl + bcl == cal) or (bcl + cal == abl) or (cal + abl == bcl)
    return g
```

```python
def points_on_circle(a, b, c, d, norm=np.linalg.norm):
    """
    Check does 4 points lie on one circle.
    
    Parameters:
    -----------
    a : float np.ndarray
    
    b : float np.ndarray
    
    c : float np.ndarray
    
    d : float np.ndarray
    
    norm : function
        Matrix or vector norm.
    
    Returns:
    g : bool
    """
    g = np.linalg.matrix_rank(np.array([a, b, c, d])) == 2
    if g:
        r1 = get_triangle_radius(a, b, c)
        r2 = get_triangle_radius(b, c, d)
        g = g and (r1 == r2)
    return g
```

```python
def data_is_general(data, norm=np.linalg.norm):
    """
    Returns True if data is in general position
    
    Parameters:
    -----------
    data : float np.ndarray
    
    norm : function
        Matrix or vector norm.
    
    Returns:
    --------
    g : bool
    """
    g = True
    for combination in itertools.combinations(data, 3):
        a, b, c = combination
        if points_on_line(a, b, c, norm):
            g = False
            break
    for combination in itertools.combinations(data, 4):
        a, b, c, d = combination
        if points_on_circle(a, b, c, d, norm):
            g = False
            break
    return g
```

# Voronoi diagram

```python
def get_middle_points(data):
    """
    Returns midles for all point pairs.
    
    Parameters:
    -----------
    data : float np.ndarray
    
    Returns:
    --------
    m : float np.ndarray
    """
    m = np.array([data for i in data])
    transpose_map = np.arange(len(m.shape))
    transpose_map[0] = 1
    transpose_map[1] = 0
    m = 0.5*(m + m.transpose(transpose_map))
    return m
```

```python
m = get_middle_points(data)
ms = [[str(m[i, j]) for i in range(len(data))] for j in range(len(data))]
ds = [str(i) for i in data]
pd.DataFrame(ms, columns=ds, index=ds)
```

```python
def get_direction_vectors(data):
    """
    Returns (pair[1] - pair[0]) for all point pairs.
    
    Parameters:
    -----------
    data : float np.ndarray
    
    Returns:
    --------
    m : float np.ndarray
    """
    m = np.array([data for i in data])
    transpose_map = np.arange(len(m.shape))
    transpose_map[0] = 1
    transpose_map[1] = 0
    m = m.transpose(transpose_map) - m
    return m
```

```python
m = get_direction_vectors(data)
ms = [[str(m[i, j]) for i in range(len(data))] for j in range(len(data))]
ds = [str(i) for i in data]
pd.DataFrame(ms, columns=ds, index=ds)
```

```python
def get_distances(data):
    """
    Returns midles for all point pairs.
    
    Parameters:
    -----------
    data : float np.ndarray
    
    Returns:
    --------
    m : float np.ndarray
    """
    m = np.array([data for i in data])
    transpose_map = np.arange(len(m.shape))
    transpose_map[0] = 1
    transpose_map[1] = 0
    m = np.linalg.norm(m - m.transpose(transpose_map), axis=2)
    return m
```

```python
m = get_distances(data)
ms = [[str(m[i, j]) for i in range(len(data))] for j in range(len(data))]
ds = [str(i) for i in data]
pd.DataFrame(ms, columns=ds, index=ds)
```

```python
def get_line_equivs(data):
    """
    Returns ...
    Ax = r
    
    Parameters:
    -----------
    data : 2-dimensional float np.ndarray
    
    Returns:
    --------
    A : float np.ndarray
    
    r : float np.ndarray
    """
    middle_points = get_middle_points(data)
    directions = get_direction_vectors(data)
    m = directions * np.array([[[1, -1]]])
    r = (m * middle_points).sum(axis=2)
    
    return m, r
```

```python
def get_midleline_intersections(data):
    """
    Returns ...
    
    Parameters:
    -----------
    data : 2-dimensional float np.ndarray
    
    Returns:
    --------
    m : dict
    """
    lines_left, lines_right = get_line_equivs(data)
    res = {}
    
    for j0 in range(len(data)):
        for j1 in range(j0+1, len(data)):
            for j2 in range(j1+1, len(data)):
                for j3 in range(j2+1, len(data)):
                    left = np.array([lines_left[j0, j1], lines_left[j2, j3]])
                    right = np.array([lines_right[j0, j1], lines_right[j2, j3]])
                    try:
                        q = np.linalg.solve(left, right)
                        res.update({(j0, j1, j2, j3) : q})
                    except np.linalg.LinAlgError:
                        pass
    return res
```

```python
midline_intersects = get_midleline_intersections(data)
midline_intersects = np.array([midline_intersects[i] for i in midline_intersects])
midline_intersects = midline_intersects[abs(midline_intersects[:, 0]) < 3.2]
midline_intersects = midline_intersects[abs(midline_intersects[:, 1]) < 3.2]

plt.figure(figsize=(12, 8))
plt.scatter(data[:, 0], data[:, 1], color='indigo', linewidth=8)
plt.scatter(midline_intersects[:, 0], midline_intersects[:, 1], color='orchid', linewidth=4)
plt.grid(True)
plt.show()
```

# Delaunay triangulation

```python
def get_hull_pairs(data):
    hull = ConvexHull(data)
    vertices = hull.vertices
    n = len(vertices)
    pairs = np.array([[vertices[i], vertices[(i+1)%n]] for i in range(n)])
    return pairs
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(*data.transpose(), color='indigo', linewidth=9)

tri = sp.spatial.Delaunay(data)
tri_simplices = tri.simplices
for triplet in tri_simplices:
    x, y = np.array([data[triplet[i%3]] for i in range(4)]).transpose()
    plt.plot(x, y, linewidth=2, color='olive')
    
for pair in get_hull_pairs(data):
    a = data[pair[0]]
    b = data[pair[1]]
    plt.plot([a[0], b[0]], [a[1], b[1]], color='orchid', linewidth=4)

plt.grid(True)
plt.show()
```
