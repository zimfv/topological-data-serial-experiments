```python
import time
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
import multiprocessing as mp
from package.filtredcells import FiltredCells
from package.tessellations import get_hexagonal_tessolation_on_torus2d


clock = time.perf_counter()

sizes = np.arange(10, 20)
experiments = 100

nCPU = mp.cpu_count()
process_ids = [np.arange(experiments)[np.arange(experiments) % nCPU == i] for i in range(nCPU)]

print('Calculating giant cycle births for {0} experiments for {1} different 2-dimensional torus splits by hexagons.'.format(experiments, len(sizes)))

bar0 = tqdm(total=len(sizes))
bar1 = tqdm(total=experiments)
bar0.set_description('Sizes checked')
bar1.set_description('Experiments')


filename = 'lab-results/lab_hexagonal_tessellation_on_torus2d/size{0}_exp{1}.csv'

def task(size, filtrations, ids, CPUid):
    fc = get_hexagonal_tessolation_on_torus2d(n=size)
    for i in ids:
        filtration = filtrations[i]
        fc.set_filtration(filtration)
        fc.initialize_complex()
        res = fc.get_giant_cycles()
        res.to_csv(filename.format(size, i))
        

for size in sizes:
    bar0.set_postfix_str('size='+str(size))
    filtrations = np.random.random([experiments, size**2])
    
    process_args = [(size, filtrations, process_ids[i], i) for i in range(nCPU)]
    process = [mp.Process(target=task, args=process_args[i]) for i in range(nCPU)]
    for i in range(nCPU):
        process[i].start()
    for i in range(nCPU):
        process[i].join()
    
    if  size != sizes[-1]:
        bar1.reset()
    bar0.update()
bar0.close()
bar1.close()
print('All sucessfuly calculated in {0: .4f} seconds.\n'.format(time.perf_counter() - clock))
```

# Analyze time

```python
import time
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
from matplotlib import colormaps as cm
from package.tessellations import get_hexagonal_tessolation_on_torus2d
```

```python
ns = range(1, 200, 4)
times = {'get_hexagonal_tessolation_on_torus2d' : [], 
         'set_filtration' : [],
         'initialize_complex' : [],
         'get_cycle_info' : [], 
         'get_giant_cycles' : []}

bar = tqdm(total=len(ns))
for n in ns:
    # get_hexagonal_tessolation_on_torus2d
    clock = time.perf_counter()
    fc = get_hexagonal_tessolation_on_torus2d(n)
    times['get_hexagonal_tessolation_on_torus2d'].append(time.perf_counter() - clock)
    # set_filtration
    filtration = np.random.random(n**2)
    clock = time.perf_counter()
    fc.set_filtration(filtration)
    times['set_filtration'].append(time.perf_counter() - clock)
    # initialize_complex
    clock = time.perf_counter()
    fc.initialize_complex()
    times['initialize_complex'].append(time.perf_counter() - clock)
    # get_cycle_info
    clock = time.perf_counter()
    fc.get_cycle_info()
    times['get_cycle_info'].append(time.perf_counter() - clock)
    # get_giant_cycles
    clock = time.perf_counter()
    fc.get_giant_cycles()
    times['get_giant_cycles'].append(time.perf_counter() - clock)
    
    bar.update()
bar.close()

for label in times:
    times[label] = np.array(times[label])
```

```python
colormap = cm['turbo']
n = len(times)
colornums = np.arange(n) / (n-1)
colors = {}
for i in range(n):
    colors.update({list(times.keys())[i] : colormap(i/(n-1))})
colors
```

```python
plt.figure(figsize=(16, 12))
for label in times:
    plt.plot(ns, times[label], linewidth=4, label=label, color=colors[label])
    plt.scatter(ns, times[label], linewidth=3, color=colors[label])
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16, 12))
plt.title('Time get_hexagonal_tessolation_on_torus2d portion of time get_cycle_info, %.')
plt.plot(ns, 100*times['get_hexagonal_tessolation_on_torus2d'] / times['get_cycle_info'], 
         linewidth=4, linestyle='-', color='steelblue')
plt.grid(True)
plt.legend()
plt.show()
```
