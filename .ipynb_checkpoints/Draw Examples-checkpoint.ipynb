```python
import numpy as np
from matplotlib import pyplot as plt
from package.tessellations import *
```

# PVLM$(Z^2, m)$

```python
m = 12
fc = get_cubical_tessellation_on_torus2d(m, filtration=np.random.random(m**2))

plt.figure(figsize=(8, 8))
fc.draw_filtration(colormap='Blues')
plt.savefig('LaTeX/pics/exampleZ2.png')
```

# PVLM$(A_2, m)$

```python
m = 6
filtration = np.random.random(m**2)
fc = get_hexagonal_tessolation_on_torus2d(m, filtration=filtration)
```

```python
figsize = (8, 8)
plt.figure(figsize=figsize)

fc.draw_filtration(colormap='Blues')

xticks = np.arange(-1, 2*m + 1)
xlabels = xticks%(2*m)
yticks = np.arange(-1, 1.5*m, 0.5)
ylabels = yticks%(1.5*m)

plt.xticks(xticks, labels=xlabels)
plt.yticks(yticks, labels=ylabels)

plt.savefig('LaTeX/pics/exampleA2.png')
```
