```python
import pickle
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
filename = 'CubesSplitingTorus2d/giant_births.pkl'
```

```python
with open(filename, 'rb') as fp:
    data = pickle.load(fp)
data = pd.DataFrame(data)
data.head()
```

```python
data['average 1st'] = [np.mean(i) for i in data['births 1st']]
data['average 2nd'] = [np.mean(i) for i in data['births 2nd']]
data['variance 1st'] = [np.var(i) for i in data['births 1st']]
data['variance 2nd'] = [np.var(i) for i in data['births 2nd']]
data['n'] = [i[0] for i in data['shape']]
data.head()
```

```python
plt.figure(figsize=(12, 9))

plt.plot(data['n'], data['average 1st'], linewidth=4, color='red', alpha=1, label='1st birth')
plt.plot(data['n'], data['average 1st'] - data['variance 1st'], linewidth=2, color='red', alpha=0.25)
plt.plot(data['n'], data['average 1st'] + data['variance 1st'], linewidth=2, color='red', alpha=0.25)

plt.plot(data['n'], data['average 2nd'], linewidth=4, color='blue', alpha=1, label='2nd birth')
plt.plot(data['n'], data['average 2nd'] - data['variance 2nd'], linewidth=2, color='blue', alpha=0.25)
plt.plot(data['n'], data['average 2nd'] + data['variance 2nd'], linewidth=2, color='blue', alpha=0.25)

plt.legend()
plt.grid(True)
plt.show()
```
