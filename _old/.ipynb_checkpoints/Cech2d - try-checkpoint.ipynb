```python
def get_distances(points, norm=np.linalg.norm):
    # 
    mat = np.concatenate([[points] for i in points])
    dis = mat - mat.transpose((1, 0, 2))
    dis = norm(dis, axis=2)
    return dis
```

```python
def get_adjacency2d(data, r):
    # returns boolean matrix size (n, n), where n is data lenth
    matrix = get_distances(data) <= 2*r
    matrix[np.eye(len(matrix), dtype=bool)] = False
    return matrix
```

```python
def get_cords(matrix):
    # returns cords of matrix, which are True
    return np.unique(np.sort(np.array(np.where(matrix)).transpose(), axis=1), axis=0)
```

```python
def draw_spmlicies1d(data, adj, ax=plt, color='blue', alpha=1, linewidth=2):
    # drawing 
    for cord in get_cords(adj):
        p0 = data[cord[0]]
        p1 = data[cord[1]]
        ax.plot([p0[0], p1[0]], [p0[1], p1[1]], color=color, alpha=alpha, linewidth=linewidth)
```

```python
def remove_negative(regions, points):
    new_regions = []
    new_points = []
    for i in range(len(regions)):
        region = regions[i]
        if len(region) > 0:
            if (region != -1).all():
                new_regions.append(np.append(regions, regions[0]))
                new_points = [points[i]]
            else:
                new_regions.append(np.concatenate([region[region.find(-1)+1:], region[:region.find(-1)]]))
                new_points = [points[i]]
    return new_regions, new_points
```

```python
def get_outer_vertices(voronoi):
    # returns dict, which containing geometry outer (real value of -1 in voronoi.ridge_dict)
    outer_vertices = {}
    for pair in voronoi.ridge_dict:
        if -1 in voronoi.ridge_dict[pair]:
            out_vector = vor.points[pair[0]] - voronoi.points[pair[1]]
            out_vector = np.array([1, -1]) * np.flip(out_vector)
            vertex = np.array(voronoi.ridge_dict[pair])
            vertex = vertex[vertex != -1][0]
            vertex = voronoi.vertices[vertex]
            neighbour = np.array(voronoi.ridge_dict[pair])
            neighbour = neighbour[neighbour != -1][0]
            ridge_vertices = np.array(voronoi.ridge_vertices)
            ridge_vertices = ridge_vertices[np.logical_and(ridge_vertices[:, 0] != -1, ridge_vertices[:, 1] != -1)]
            ridge_vertices = ridge_vertices[np.logical_or(ridge_vertices[:, 0] == neighbour, ridge_vertices[:, 1] == neighbour)]
            neighbour = ridge_vertices[ridge_vertices != neighbour][0]
            neighbour = voronoi.vertices[neighbour]
            in_vector = neighbour - vertex
            if np.dot(out_vector, in_vector) > 0:
                out_vector *= -1
            out_vertex = vertex + out_vector
            outer_vertices.update({pair : out_vertex})
    return outer_vertices
```

```python
def get_ridge_dict_cords(voronoi):
    # returns same dict as voronoi.ridge_dict, but coordinates of vertices, not indices
    outer_vertices = get_outer_vertices(voronoi)
    ridge_dict_cords = {}
    for pair in voronoi.ridge_dict:
        verts = voronoi.ridge_dict[pair]
        if not(-1 in verts):
            ridge_dict_cords.update({pair : list(voronoi.vertices[verts])})
        else:
            inner = np.array(verts)
            inner = inner[inner != -1][0]
            inner = voronoi.vertices[inner]
            ridge_dict_cords.update({pair : [inner, outer_vertices[pair]]})
    return ridge_dict_cords
```

```python
def line_equation(a0, a1):
    # returns A[i] and b[i], where i's string is line (a0, a1) in linear system Ax = b
    a = np.array([a1[1]-a0[1], -a1[0]+a0[0]])
    b = (a0*a).sum()
    return a, b


def line_function(a0, a1):
    # returns function f(x) = A*x[0] + B*x[1] + C, where f(x)=0 is equation for line (a0, a1)
    (A, B), C = line_equation(a0, a1)
    C = -C
    return lambda x: A*x[0] + B*x[1] + C


def lines_intersection(a0, a1, b0, b1):
    # return intersection of lines (a0, a1) and (b0, b1)
    eq_a = line_equation(a0, a1)
    eq_b = line_equation(b0, b1)
    matrix = np.array([eq_a[0], eq_b[0]])
    r = np.array([eq_a[1], eq_b[1]])
    return np.linalg.solve(matrix, r)
```

```python
def point_in_segment(p, a, b):
    # returns True if point p, which is lying on line ab, is in segment ab
    if (a[0] < p[0]) and (p[0] < b[0]):
        return True
    if (b[0] < p[0]) and (p[0] < a[0]):
        return True
    if (a[1] < p[1]) and (p[1] < b[1]):
        return True
    if (b[1] < p[1]) and (p[1] < a[1]):
        return True
    return False
```

```python
def get_projection(a0, a1, b):
    # returns projection of point b to line (a0, a1)
    l = np.array(a1) - np.array(a0)
    n = np.array([l[1], -l[0]])
    return lines_intersection(a0, a1, b, b+n)


def get_distance_line_point(a0, a1, b):
    # returns distance between point b and line (a0, a1)
    return np.linalg.norm(b - get_projection(a0, a1, b))
```

```python
def get_radiuses_for_cycles1d(voronoi):
    # 
    rdc = get_ridge_dict_cords(voronoi)
    radiuses = {}
    for pair in rdc:
        point0 = voronoi.points[pair[0]]
        point1 = voronoi.points[pair[1]]
        vertex0 = rdc[pair][0]
        vertex1 = rdc[pair][1]
        intersection = lines_intersection(point0, point1, vertex0, vertex1)
        if point_in_segment(intersection, vertex0, vertex1):
            #radius = get_distance_line_point(vertex0, vertex1, point0)
            #radius = get_distance_line_point(vertex0, vertex1, point1)
            radius = 0.5*np.linalg.norm(point0 - point1)
            radiuses.update({pair: radius})
    return radiuses
```

```python
def get_parties_for_cycles2d(voronoi, filter_abtuse=True):
    # 
    parties = {}
    for vertex in range(len(voronoi.vertices)):
        party = np.array([], dtype=int)
        for pair in voronoi.ridge_dict:
            ridge = voronoi.ridge_dict[pair]
            if vertex in ridge:
                party = np.concatenate([party, pair])
        party = np.unique(party)
        g = True
        if (len(party) == 3) and filter_abtuse:
            a, b, c = voronoi.points[party]
            if np.dot(b-a, c-a) < 0:
                g = False
            if np.dot(a-b, c-b) < 0:
                g = False
            if np.dot(a-c, b-c) < 0:
                g = False
        if g:  
            parties.update({vertex : party})
    return parties
```

```python
def get_phase_radiuses(voronoi):
    # 
    phases = [{'radius': 0, 'cycle dimension' : 0}]
    radiuses1d = get_radiuses_for_cycles1d(voronoi)
    for pair in radiuses1d:
        phases.append({'radius' : radiuses1d[pair], 
                       'cycle dimension' : 1, 
                       'point indicies' : pair, 
                       'intersection' : 0.5*(voronoi.points[pair[0]] + voronoi.points[pair[1]])})
    parties2d = get_parties_for_cycles2d(voronoi)
    for vertex in parties2d:
        intersection = voronoi.vertices[vertex]
        radius = np.linalg.norm(intersection - voronoi.points[parties2d[vertex][0]])
        phases.append({'radius' : radius, 
                       'cycle dimension' : 2, 
                       'point indicies' : tuple(parties2d[vertex]), 
                       'intersection' : intersection})
    phases = pd.DataFrame(phases)
    phases = phases.sort_values(by='radius')
    phases = phases.reset_index(drop=True)
    return phases
```

```python
class Cech2d(Voronoi):
    def get_phases(self):
        return get_phase_radiuses(self)
    
    def get_complex(self, radius):
        phases = self.get_phases()
        phases = phases[phases['radius'] < radius]
        ridges0 = np.arange(len(self.points))
        ridges1 = phases[phases['cycle dimension'] == 1]['point indicies'].values
        ridges2 = phases[phases['cycle dimension'] == 2]['point indicies'].values
        return {0: ridges0, 1: ridges1, 2: ridges2}
    
    def draw_complex(self, radius, ax=plt,
                     color0='steelblue', alpha0=1.0, width0=6, 
                     color1='steelblue', alpha1=1.0, width1=3, 
                     color2='steelblue', alpha2=0.4):
        komplex = self.get_complex(radius)
        for ridge in komplex[2]:
            points = self.points[np.array(ridge)]
            ax.fill(points[:, 0], points[:, 1], color=color2, alpha=alpha2)
        for ridge in komplex[1]:
            points = self.points[np.array(ridge)]
            ax.plot(points[:, 0], points[:, 1], color=color1, alpha=alpha1, linewidth=width1)
        ax.scatter(self.points[:, 0], self.points[:, 1], color=color0, alpha=alpha0, linewidth=width0)
```
